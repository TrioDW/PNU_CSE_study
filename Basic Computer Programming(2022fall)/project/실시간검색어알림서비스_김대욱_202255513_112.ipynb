```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["Blockchain"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')
```

```python
df = pytrends.trending_searches(pn='united_states') # trending searches in real time for United States
df
```

```python
import googletrans

translator = googletrans.Translator()

for i in range(20) :
  result = translator.translate(df.iloc[i, 0], dest='ko')
  df.iloc[i, 0] = result.text

df
```
